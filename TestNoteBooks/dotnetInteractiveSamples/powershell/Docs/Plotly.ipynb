```powershell
$dnld = Invoke-RestMethod https://raw.githubusercontent.com/plotly/datasets/master/mesh_dataset.txt
$data = $dnld | ConvertFrom-Csv -Delimiter " " -Header X, Y, Z
```

```powershell
function DoGraph {
    param(
        $type,
        [hashtable]$h,
        $title
    )

    $target = "Graph.$($type)"
    $series = new-object $target
    
    $h.keys | % {
        $series.$_ = $h.$_
    }

    $series | New-PlotlyChart -Title "Mesh Dataset" | Out-Display

}

$type = 'Mesh3d'
$title = "Test"
$h = @{
    x         = $data.X
    y         = $data.Y
    z         = $data.Z
    color     = 'cyan'
    opacity   = 0.4
    alphahull = 5
}

DoGraph $type $h $title

return

$target = "Graph.$($type)"
$series = new-object $target
$series.x=$data.x
$series.y=$data.y
$series.z=$data.z
$series.color   = 'lightpink'
$series.opacity = 0.50

$series | New-PlotlyChart -Title "Mesh Dataset" | Out-Display

#$openSeries = [Graph.Scatter]::new()
#$openSeries.name = "Open"
#$openSeries.x = @(1, 2, 3, 4)
#$openSeries.y = @(10, 15, 13, 17)

#$closeSeries = [Graph.Scatter]::new()
#$closeSeries.name = "Close"
#$closeSeries.x = @(2, 3, 4, 5)
#$closeSeries.y = @(16, 5, 11, 9)

#$chart = @($openSeries, $closeSeries) | New-PlotlyChart -Title "Open vs Close"
#Out-Display $chart

return 
#$series = [$target]@{
$series = [Graph.Mesh3d]@{
    x       = $data.X
    y       = $data.Y
    z       = $data.Z
    color   = 'lightpink'
    opacity = 0.50
}

$series | New-PlotlyChart -Title "Mesh Dataset" | Out-Display
```
