```powershell
ipmo ..\PSKit.psd1 -force

function Show-ConfirmedCases {
    param(
        [Parameter(ValueFromPipeline)]
        $location
    )
    
    Begin {
        $baseUrl = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data'
        $url = "$baseUrl/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
    }
    
    Process {
        $subset = (read-csv $url).Where({$_.'Province/State' -eq $location})
     
         (Invoke-Pivot -targetData $subset -columnPattern '^3/\d{2}/*').
             Plot('x','y',"confirmed cases - $location")
    }
}

'Texas' | Show-ConfirmedCases
```

```powershell
$baseUrl = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data'
$url = "$baseUrl/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"

$subset = (read-csv $url).Where({$_.'Province/State' -match "New York$|Washington$|Texas|Florida"}) | Sort 'Province/State'

#$subset | ctmt

$d = New-Dimension -targetData $subset -columnPattern '3/\d{2}/20' -dimension 'Province/State' -measureName Date

# transform the data

"Florida: $($d.florida)"

$d|ctmt
```
