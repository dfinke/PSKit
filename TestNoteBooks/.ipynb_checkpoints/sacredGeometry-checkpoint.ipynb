```python
from numpy import sin, cos, linspace, pi
import matplotlib.pyplot as plt
```

```python
def draw(p, q, n=200):
    xs = linspace(0, 2*pi, n)
    plt.axes().set_aspect(1)
    plt.axis('off')
    for x in xs:
        plt.plot([cos(p*x), cos(q*x)],[sin(p*x), sin(q*x)], "b-", alpha=.4)
    plt.show()
    plt.close()

draw(4, 3)
# draw(4, 1)
# draw(7, 3)
#draw(9, 4)
```

```python
def draw(n=10):
    pi=3.141592653589793
    r=.5

    s = 2*pi/n

    plt.axes().set_aspect(1)
    plt.axis('off')

    for i in range(n-1):
        for j in range(i+1,n):
            p1=r*cos(s*i)
            p2=r*sin(s*i)
            p3=r*cos(s*j)
            p4=r*sin(s*j)
            plt.plot([p1, p3],[p2,p4], "b-", alpha=.4)

    plt.show()
    plt.close()
    
draw()
```
